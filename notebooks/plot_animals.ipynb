```python
import matplotlib.pyplot as plt 
import numpy as np
```

```python

ACC = []
for i in range(50):
    ACC.append(np.loadtxt('../results/ANIMALS/SGD/accuracy' + str(i)))

ACC = np.array(ACC)

# Set up the plot
plt.figure(figsize=(6, 4))

# Define the number of metrics, based on the third dimension of ACC (assuming ACC has shape [10, x, 5])
num_metrics = ACC.shape[2]

# Plot mean ± standard deviation for each metric
colors = ['C0', 'C1', 'C2', 'C3', 'C4']  # Define colors for each metric line

for i in range(num_metrics):
    mean_acc = np.mean(ACC[:, :, i], axis=0)
    std_acc = np.std(ACC[:, :, i], axis=0)
    
    # Plot the mean line
    plt.plot(mean_acc,color=colors[i], label=f'Task {i+1}')
    
    # Plot the shaded area for mean ± std
    plt.fill_between(range(mean_acc.shape[0]),
                     mean_acc - std_acc,
                     mean_acc + std_acc,
                     color=colors[i],
                     alpha=0.1)

# Label the axes
plt.xlabel('Epochs', fontsize=18)
plt.ylabel('Accuracy', fontsize=18)

# Set tick parameters for better readability
plt.xticks([0,5,10,15,20],fontsize=16)
plt.yticks(fontsize=16)
plt.ylim([0.5, 1])

# Add legend with larger font size
plt.legend(fontsize=16)

# Save the figure
plt.savefig("accuracy_det.svg", format='svg')

# Show the plot
plt.show()

```

```python
aleatoric=[]
aleatoric_un=[]
for i in range(50):
    aleatoric.append(np.loadtxt('../results/ANIMALS/SGD/aleatoric'+str(i)))
    aleatoric_un.append(np.loadtxt('../results/ANIMALS/SGD/aleatoric_un'+str(i)))

aleatoric=np.array(aleatoric)
aleatoric_un=np.array(aleatoric_un)
# Set figure size
plt.figure(figsize=(6, 4))

# Plot histograms with transparency
plt.hist(aleatoric.flatten(), density=True, bins=20, color='C0', alpha=0.6,label='In distribution')
plt.hist(aleatoric_un.flatten(), density=True, bins=20, color='C3', alpha=0.6,label='Out of distribution')

# Set labels with larger font size
plt.xlabel('Aleatoric uncertainty', fontsize=18)
plt.ylabel('Density', fontsize=18)

# Set tick parameters for better readability
plt.xticks([0,0.5,1,1.5],fontsize=16)
plt.yticks([0,3,6,9],fontsize=16)
plt.legend(fontsize=16)

# Add legend with larger font size
# plt.legend(fontsize=16)

# Display the plot
plt.savefig("uncertainty_dist_det.svg", format='svg')

plt.show()

```
